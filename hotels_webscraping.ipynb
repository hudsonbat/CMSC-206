```python
pages = range(0,2490,30)
```

```python
links = []

for page in pages:
    if page == 0:
        html = requests.get('https://www.tripadvisor.in/Hotels-g187147-Paris_Ile_de_France-Hotels.html',headers=headers)
        bsobj = soup(html.content,'lxml')
        for review in bsobj.findAll('a',{'class':'review_count'}):
            a = review['href']
            a = 'https://www.tripadvisor.in'+ a
            links.append(a)
    else:
        html = requests.get('https://www.tripadvisor.in/Hotels-g187147-oa' + str(page) + '-Paris_Ile_de_France-Hotels.html',headers=headers)
        print('https://www.tripadvisor.in/Hotels-g187147-oa' + str(page) + '-Paris_Ile_de_France-Hotels.html')
        bsobj = soup(html.content,'lxml')
        for review in bsobj.findAll('a',{'class':'review_count'}):
            a = review['href']
            a = 'https://www.tripadvisor.in'+ a
            links.append(a)
        sleep(randint(10,20))
```

```python
for link in links:
    html2 = requests.get(link,headers=headers)
    sleep(randint(1,5))
    bsobj2 = soup(html2.content,'lxml')
    for r in bsobj2.findAll('q'):
        try:
            review_list.append(r.span.text.strip())
        except:
            review_list.append('')
    reviews[link]= review_list
    review_list = []
```

```python
import re

data = pd.read_csv('hotel_reviews_final2.csv',index_col=0)
data.head(20)
```

```python
name = data['Hotel_Name'][0]
name
```

```python
new_name = ''
index = 0

index = name.find('-Reviews-')
index += 9
new_name += name[index:-13]
new_name
```

```python
for i in range(len(data)):
    name = ''
    index = data['Hotel_Name'][i].find('-Reviews-')
    index += 9
    name += data['Hotel_Name'][i][index:-13]
    data['Hotel_Name'][i] = name
    index = 0
    
data.head()
```

```python
for i in range(len(data)):
    rating = data['Rating'][i]
    rating2 = re.sub('\D','',rating)
    rating3 = float(rating2[0])
    data['Rating'][i] = rating3

data.head()
```

```python
for i in range(len(data)):
    room = data['Number_Of_Rooms'][i]
    room2 = re.sub('\D','',room)
    room3 = int(room2)
    data['Number_Of_Rooms'][i] = room3

data.head()
```

```python
data['Low_Price_Room'] = 0
data['High_Price_Room'] = 0

for i in range(len(data)):
    price = data['Price_Range'][i]
    prices = price.split('-')
    try:
        data['Low_Price_Room'][i] = int(re.sub('\D','',prices[0]))
        data['High_Price_Room'][i] = int(re.sub('\D','',prices[1]))
    except:
        continue
        
data.head()
```

```python
data['B&B_Ranking'] = 0
data['Hotel_Ranking'] = 0
data['Spec_Lodge_Ranking'] = 0
```

```python
for i in range(len(data)):
    if 'hotels' in data['Ranking'][i]:
        data['Hotel_Ranking'][i] = int(re.search('[0-9]+', data['Ranking'][i]).group())
    elif 'B&Bs' in data['Ranking'][i]:
        data['B&B_Ranking'][i] = int(re.search('[0-9]+', data['Ranking'][i]).group())
    elif 'lodging' in data['Ranking'][i]:
        data['Spec_Lodge_Ranking'][i] = int(re.search('[0-9]+', data['Ranking'][i]).group())
```
